```python
#install libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
cust_ord_prod = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'cust_orders_products_all.pkl'))
```

# Data Visualization

```python
#create a histogram of "order_hour_of_day" column
hist = cust_ord_prod['order_hour_of_day'].plot.hist(bins = 24, x='Hour of Day', y ='Number of Orders', color = 'purple' )
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.title('Frequency of Hour of Day Ordered')
```

The histogram illustrates that a majority of shopping takes place in the late morning/afternoon hours between 9 am and 4 pm.

```python
#create a bar chart from the "loyalty_flag" column
bar = cust_ord_prod['loyalty_flag'].value_counts().plot.bar(color = ['darkgreen', 'green', 'limegreen'])
plt.title('Loyalty Tag Frequency')
```

Regular customers are the most frequent shoppers, followed by loyal customers then new customers

# Line Graph: Age vs Number of Dependents

```python
#create a subset for line chart exploring the connections between age and number of dependents
df_1 = small[['Age', 'n_dependants']]
```

```python
#create a line chart exploring the connections between age and number of dependents
line_2 = sns.lineplot(data = df_1, x = 'Age',y = 'n_dependants')
```

There is no discernible connection between age and number of dependents in the sample. A majority of people sampled have between 1 and 2 dependents.

# Scatterplot: Age vs Spending Power

```python
scat1 = sns.scatterplot(x = 'Age', y = 'income',data = small, alpha=0.1)
scat1.set_ylim(0, 200000)
```

```python
#plot a scatterplot to explore whether there’s a connection between age and spending power (income)
scat = sns.scatterplot(x = 'Age', y = 'income',data = small, alpha=0.1)
scat.set_ylim(200000, 600000)
```

Most income is concentrated between 25,000 and 200,000 but there is a higher concentration of income over 200,000 in the age range between 40 and 80.
