# Create a loyalty flag using loc()

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
#create categories for 'loyalty_flag' column
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#find frequency of "loyalty_flag" column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(20)
```

```python
#compare spending habits of three types of customers
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'max', 'min',]})
```

```python
ords_prods_merge.describe()
```

# Create a spending flag

```python
#create a spending flag based on average price across all orders
ords_prods_merge['user_avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
#create categories for 'user_avg_price' column
ords_prods_merge.loc[ords_prods_merge['user_avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#create categories for 'user_avg_price' column
ords_prods_merge.loc[ords_prods_merge['user_avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#find frequency of "spending_flag" column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# Create a frequency flag

```python
#create a frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column
ords_prods_merge['med_order_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
#create categories for the 'order_frequency_flag' column
ords_prods_merge.loc[ords_prods_merge['med_order_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
#create categories for the 'order_frequency_flag' column
ords_prods_merge.loc[(ords_prods_merge['med_order_freq'] > 10) & (ords_prods_merge['med_order_freq'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
#create categories for the 'order_frequency_flag' column
ords_prods_merge.loc[ords_prods_merge['med_order_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
#find frequency of "order_freq_flag" column
ords_prods_merge['order_freq_flag'].value_counts(dropna = False)
```

```python
# Export altered dataframe to pkl

ords_prods_merge.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_agg.pkl'))
```
